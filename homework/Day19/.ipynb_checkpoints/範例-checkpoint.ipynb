範例目標:<br>
1. 實做繪製圖表
2. 藉由圖表對資料有初步認識，並解釋圖表意義

範例重點:<br>
1. 每一種圖表都有相對應用的情境，分辨情境並繪製對應圖表
2. 圖像化展現，方便對於資料有初步的理解

```python
import pandas as pd
import numpy as np
```

```python
#單一折線圖     #random.randn(200) size = 200 隨機得到200個數字，可能是負的或正的，可能大於一或小於
ts = pd.Series(np.random.randn(200), index=pd.date_range('1/1/2020', periods=200))
ts = ts.cumsum()  #用於計算給定軸上數組元素的累積和
print(ts)
ts.plot()
```

```python
#多個折線圖同時畫出               #200x3
df = pd.DataFrame(np.random.randn(200, 3), index=pd.date_range('1/1/2020', periods=200), columns=["A","B","C"])
df = df.cumsum()
print(df)
df.plot();
```

```python
#圓餅圖                   #size = 4
df = pd.Series(np.random.rand(4), index=["A","B","C","D"],name="title")
print(df)
df.plot.pie()
```

```python
#長條圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.bar();
```

```python
df.plot.bar(stacked=True); #堆疊A、B
```

```python
#箱型圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.boxplot()
```

```python
#散佈圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.scatter(x='A', y='B')
```
