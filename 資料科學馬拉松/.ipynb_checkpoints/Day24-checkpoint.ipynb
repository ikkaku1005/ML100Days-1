# 利用 Bokeh 與 Python 製作網頁互動視覺化
設定資料與輸出檔案 output_file(“out.html”) 利用 Bokeh 繪製圖表 p = figure() 開啟產生的 HTML 檔 ( HTML + JavaScript ，自動生成 ) show(p)

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import widgets
from bokeh.io import output_notebook
# 讓網頁直接輸出在NOTEBOOK
output_file("out.html")
p = figure()
p.line([1,2,3,4,5],[5,4,3,2,1])
#劃一個調線:(1,5)(2,4)(3,3)(4,2)(5,1) 斜率為-1的line
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
# 讓網頁直接輸出在NOTEBOOK
output_notebook()
#創造一個畫布
p = figure(plot_width = 300, plot_height = 300, tools = "pan, reset, save")

# 添加圓點的分佈圖
p.circle([1, 2.5, 3, 2], [2, 3, 1, 1.5], size=20, color="navy", alpha=0.5)
# p.circle (x/y 座標, size, color, and 透明度)
output_file("foo.html")
show(p)
```

# 配置工具欄

```python
from bokeh.plotting import figure, output_file, show

output_file("toolbar.html")
# 新增工具欄, 
p = figure(plot_width=400,  plot_height = 400,
           title = None, toolbar_location="below",  #位置在繪製圖區域下方
           toolbar_sticky=False)
# 工具欄位置與軸發生衝突，在這種情況下，將toolbar_sticky選項設置為False將工具欄移動到繪製軸的區域之外。
p.circle([1, 2, 3, 4, 5], [2, 5, 8, 2, 7], size=10)

show(p)
```

# 繪製股市資料

```python
import bokeh.sampledata
bokeh.sampledata.download()
```

```python
import numpy as np
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.stocks import AAPL
# 準備資料
aapl = np.array( AAPL['adj_close'] )
aapl_dates = np.array( AAPL['date'], dtype=np.datetime64 )

#設定視窗
window_size = 30
window = np.ones(window_size) / float(window_size)
aapl_avg = np.convolve(aapl, window, 'same')
#Numpy convolve（）方法用於返回兩個一維向量的離散線性成交量積。成交量積是什麼意思？

# 輸出靜態 HTML file
output_file("stocks.html", title="stocks.py example")

# create a new plot with a datetime axis type
p = figure( plot_width=800, plot_height = 350, x_axis_type="datetime")

# add renderers
p.circle(aapl_dates, aapl, size=4, color='darkgrey', alpha=0.2, legend_label='close')
p.line(aapl_dates, aapl_avg, color='navy', legend_label='avg')

# NEW: customize by setting attributes
p.title.text = "AAPL One-Month Average"
p.legend.location = "top_left"
p.grid.grid_line_alpha = 0
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Price'
p.ygrid.band_fill_color = "olive"
p.ygrid.band_fill_alpha = 0.1

# show the results
show(p)
```
