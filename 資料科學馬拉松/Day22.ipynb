目標: 使用 Seaborn 自帶的dataset, 利用 PANDAS 處理來繪製資料集的可是畫圖表

重點: 不同型態的類別要套用的對應圖像是不一樣的

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset("iris")
df
```

```python
df.info()
```

```python
df.describe()  #提供各種數據
```

```python
sns.boxplot(data = df, orient = 'h') #orient = 'h' 相形圖方向為水平
```

```python
sns.stripplot(x = "species", y = "petal_length", data = df)
#jitter = True (默認) 資料每個都看得到
```

```python
sns.stripplot(x = "species", y = "petal_length", data = df, jitter=False)
#jitter : 若資料集太集中，看起來會變成一直線，這個參數能對他做些干擾
#jitter=False  表示不讓資料太分散 
```

```python
sns.boxplot(x = "species", y = "petal_length", data = df)
```

```python
g = sns.pairplot(df , hue = "species", diag_kind = "kde", kind ="scatter", palette = "husl" )
#diag_kind : 用於控制對角線上的圖形 ， "kde" (曲線圖)
#kind：用於控制非對角線 ( 左上至右下 ) 上的圖形

g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw = 3)
```
